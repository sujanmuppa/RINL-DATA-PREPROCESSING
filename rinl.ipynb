```python
import pandas as pd
import numpy as np
```

```python
path = "data/1to6.xlsx"
df1= pd.read_excel(path, sheet_name="1")
df2= pd.read_excel(path, sheet_name="2")
df3= pd.read_excel(path, sheet_name="3")
df4= pd.read_excel(path, sheet_name="4")
df5= pd.read_excel(path, sheet_name="5")
df6= pd.read_excel(path, sheet_name="6")

```

```python
values = df1.values
tests = set()
for i in values:
    tests.add(i[5])
print(tests)
print(len(tests))
```

```python
data_dictionary = {}
error = 0
for i in values:
    if (i[1],i[3]) not in data_dictionary:
        data_dictionary[(i[1],i[3])] = {}
        if i[5] not in data_dictionary[(i[1],i[3])]:
            data_dictionary[(i[1],i[3])][i[5]] = i[6]
        else:
            if i[6] != "Not being done" or i[6] != "Nil":
                if data_dictionary[(i[1],i[3])][i[5]] == "Not being done" or data_dictionary[(i[1],i[3])][i[5]] == "Nil":
                    data_dictionary[(i[1],i[3])][i[5]] = i[6]
    else:
        if i[5] not in data_dictionary[(i[1],i[3])]:
            data_dictionary[(i[1],i[3])][i[5]] = i[6]
        else:
            if i[6] != "Not being done" or i[6] != "Nil":
                if data_dictionary[(i[1],i[3])][i[5]] == "Not being done" or data_dictionary[(i[1],i[3])][i[5]] == "Nil":
                    data_dictionary[(i[1],i[3])][i[5]] = i[6]
for i in data_dictionary:
    print(i, data_dictionary[i])
```

```python
data_reform = []
for i in data_dictionary.keys():
    temp = []
    temp.append(i[0])
    temp.append(i[1])
    for j in tests:
        if j in data_dictionary[i]:
            temp.append(data_dictionary[i][j])
        else:
            temp.append("Not Recorded")
    data_reform.append(temp)
print(data_reform)
```
